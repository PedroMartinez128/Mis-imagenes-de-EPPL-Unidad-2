#### 📘 Redacción y Contexto del Problema
En esta práctica de programación dirigida a nivel preparatoria, aprenderás a traducir conceptos geométricos básicos a un algoritmo funcional en Python. El objetivo es crear un programa interactivo que permita calcular tres magnitudes fundamentales relacionadas con figuras circulares: el área de un círculo, su perímetro (circunferencia) y el volumen de una esfera. Estas fórmulas son la base de múltiples aplicaciones en física, ingeniería y diseño gráfico.


---
1. Area del circulo: A = π · r²
2. Perimetro del circulo: P = 2 · π · r
3. Volumen de una esfera: V = (4/3) · π · r³

🧠 Estructura y Lógica del Código
El programa sigue un enfoque procedural (sin uso de clases ni objetos) y se organiza en cuatro funciones principales, cumpliendo estrictamente con los requisitos solicitados:

🔹 areacirculo()
Solicita el radio al usuario mediante input().
Aplica la fórmula math.pi * radio ** 2.
Imprime el resultado con formato .2f y la unidad unidades².

🔹 perimetrocirculo()
Pide el radio.
Calcula 2 * math.pi * radio.
Muestra el perímetro en unidades.

🔹 volumenesfera()
Solicita el radio de la esfera.
Ejecuta (4/3) * math.pi * radio ** 3.
Presenta el volumen en unidades³.

🔹 menu()
Utiliza un bucle while True: para mantener el programa activo.
Muestra un encabezado decorado con emojis y las 4 opciones numéricas.
Captura la elección del usuario y la evalúa con if / elif / else:
1 → Llama a areacirculo()
2 → Llama a perimetrocirculo()
3 → Llama a volumenesfera()
4 → Imprime mensaje de despedida y ejecuta break para salir del bucle.
Cualquier otro valor → Muestra advertencia y vuelve al menú

```python
# 🐍 Calculadora Geométrica Básica - Nivel Principiante
# Autor: Pedro Martinez
# Formulas utilizadas :
# 1. Area del circulo: A = π · r²
# 2. Perimetro del circulo: P = 2 · π · r
# 3.Volumen de una esfera: V = (4/3) · π · r
# Fecha: 2026

import math  # Módulo estándar para usar el valor preciso de π

def areacirculo():
    """Calcula el área de un círculo"""
    radio = float(input("📏 Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"🟢 El área del círculo es: {area:.2f} unidades²\n")

def perimetrocirculo():
    """Calcula el perímetro (circunferencia) de un círculo"""
    radio = float(input("📏 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"🔵 El perímetro del círculo es: {perimetro:.2f} unidades\n")

def volumenesfera():
    """Calcula el volumen de una esfera"""
    radio = float(input("📏 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"🟣 El volumen de la esfera es: {volumen:.2f} unidades³\n")

def menu():
    """Muestra el menú principal y controla el flujo del programa"""
    while True:
        print("=" * 50)
        print("      🧮 MENÚ PRINCIPAL 🧮")
        print("=" * 50)
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")

        opcion = input("👉 Selecciona una opción: ")

        if opcion == "1":
            areacirculo()
        elif opcion == "2":
            perimetrocirculo()
        elif opcion == "3":
            volumenesfera()
        elif opcion == "4":
            print("\n👋 ¡Hasta luego! Gracias por usar el programa. 🚀")
            print("Programa realizado por Pedro Martinez NL = 34")
            break
        else:
            print("⚠️ Opción no válida. Por favor, elige un número del 1 al 4.\n")

# 🟢 Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
